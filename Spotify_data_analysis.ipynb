```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks = pd.read_csv('tracks.csv')
df_tracks.head()
```

```python
# Checking for null values
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
#Finding 10 least popular artist
sorted_df=df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
#Finding some descriptive statistics for numerical values in our columns
df_tracks.describe().transpose()
```

```python
#top 10 songs with popularity greater than 90
most_popular=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
#Setting our index as release date column
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
#Checking the artist at a particular row
df_tracks[["artists"]].iloc[18]
```

```python
#Converting the duration of the songs from milliseconds to seconds
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
#Correlation map between different attributes
corr_df=df_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("Correlation between different variables in a form of a heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#Sample a data (0.4 percent) of total data and create two regression plots using this data
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#Creting a regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="Loudness v/s Energy relationship")
```

```python
#Creting a regression plot between loudness and energy
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="b").set(title="Popularity v/s acousticness relationship")
```

```python
#Now, I need the release year from our release date. I ll create a new column
#for release year from release date
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#Now I want to create a distribution plot to visualization total number of songs in 
#each year since 1922
#PS: we need seaborn to be of version 0.11.0
sns.displot(years,discrete=True,aspect=2,height=5,kind="hist").set(title="Number of songs per year")
```

```python
#Duration of songs over the years
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Year v/s song duration")
plt.xticks(rotation=90)
```

```python
#Creating a line plot to determine the average duration of songs over the years
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year v/s Duration")
plt.xticks(rotation=60)
```
